## Problem 1 — Write Your First Function  *(10 points)*

Turn a repeated calculation into a reusable function — the whole point of this week.

### What to do
1. Define a function named `order_total(price, quantity, tax_rate=0.0825)` that:
   - computes the subtotal as `price * quantity`,
   - adds tax (`subtotal * tax_rate`),
   - **returns** the final total **rounded to 2 decimals** (use `round(value, 2)`).
2. Notice `tax_rate` has a **default value** of `0.0825` (Texas sales tax). A caller can leave it out or override it.
3. Call your function **three times** and print each result with a label:
   - `order_total(4.50, 3)` — 3 lattes at the default tax rate,
   - `order_total(3.25, 2, 0.0)` — 2 muffins with **no tax** (override the default),
   - one call of your own choosing.
4. Add a **comment** explaining the difference between a function's **parameter** and the **argument** you pass when you call it.

### Expected output for the first two calls
Your exact labels may be different, but the numbers should match:
```text
order_total(4.50, 3) -> 14.61
order_total(3.25, 2, 0.0) -> 6.5 or 6.50
```

Note: `round(value, 2)` may display `6.5` instead of `6.50`. Either is acceptable.

### Hints
- A function starts with `def name(parameters):` and uses `return` to send a value back.
- A default argument lets you call the function with or without that value: `order_total(4.50, 3)` vs `order_total(4.50, 3, 0.0)`.

````python
# Name: Fathima Mohammed
# Problem 1: This program is about defining a function: parameters, default argument, return

# Defining functions:
def order_total(price, quantity, tax_rate = 0.0825):
  subtotal = price * quantity
  tax = subtotal * tax_rate
  total = subtotal + tax
  return round (total, 2) # rounding numbers to 2 decimal points

# Calling function:
latte_total = order_total (4.50, 3)
print ("order_total (4.50, 3) ->", latte_total)

muffin_total = order_total (3.25, 2, 0.0)
print ("order_total (3.25, 2, 0.0) ->", muffin_total)

pastry_total = order_total (5.00, 2)
print ("order_total (5.00, 2, 0.0825) ->", pastry_total)

# parameter: variable name inside the function
# argument: actual value passed into the function call.


````

## Problem 2 — Pair Up Data with `zip()`  *(10 points)*

You have two parallel lists — names and their quiz scores — and you want to work with them together.

### What to do
The starter lists are provided:
```python
names = ["Maria", "Devon", "Aisha", "Carlos"]
scores = [88, 72, 95, 60]
```
1. Use `zip()` in a `for` loop to print each name with its score and a pass/review label, like:
   ```
   Maria: 88 -> pass
   Devon: 72 -> pass
   Aisha: 95 -> pass
   Carlos: 60 -> review
   ```
   A score of **70 or higher** is `pass`; below 70 is `review`.
2. Build a list of `(name, score)` pairs by wrapping your zip in `list(...)`, store it in `paired`, and print it.
3. Use **one built-in function** of your choice on `scores` — for example `max(scores)`, `min(scores)`, `sum(scores)`, or `len(scores)` — and print the result with a label.
4. Add a **comment**: what happens if the two lists passed to `zip()` have **different lengths**? (One sentence.)

### Hint
`zip(names, scores)` yields pairs you can unpack directly: `for name, score in zip(names, scores):`.

````python
# Name: Fathima Mohammed
# Problem 2: This program is about using zip() to combine lists + a built-in function

names = ["Maria", "Devon", "Aisha", "Carlos"]
scores = [88, 72, 95, 60]

for name, score in zip(names, scores):
  if score >= 70:
    print(f"{name}: {score} -> pass") # above or equal 70 then pass
  else:
    print (f"{name}: {score} -> review") # below 70 fail

paired = list(zip(names, scores)) # Paired with name and its score.
print ("Paired:", paired)


# Ussing one built in function:
total_score = sum(scores)
print ("Total score:", total_score)

# If two lists are passed to zip() have different lengths,
# zip() stops at the shortest list and ignores extra items.


````

## Problem 3 — Recursion and Lambda  *(10 points)*

Two compact tools from this week: a function that calls itself, and a one-line function.

### Part A — Recursion
Assume `n` is a non-negative integer.

1. Define a recursive function `factorial(n)` that returns `n!` (n factorial):
   - **Base case:** when `n` is `0` or `1`, return `1`.
   - **Recursive case:** otherwise return `n * factorial(n - 1)`.
2. Print `factorial(5)` (should be `120`) and `factorial(0)` (should be `1`).
3. Add a **comment**: why does a recursive function need a **base case**? Name what happens without one. (One sentence — e.g., it keeps calling itself and eventually raises a `RecursionError`.)

### Part B — Lambda
4. Create a lambda named `double` that takes one number and returns it times 2. Print `double(7)`.
5. Use `map()` with a lambda to double every number in `nums = [1, 2, 3, 4, 5]`, convert the result to a list, and print it (expected: `[2, 4, 6, 8, 10]`).

### Hints
- A lambda is a small anonymous function: `square = lambda x: x * x`.
- `list(map(lambda x: x * 2, nums))` applies the lambda to each item.

````python
# Name: Fathima Mohammed
# Problem 3: This program is about Recursion and a lambda expression

# Part 1: Recursion
nums = [1, 2, 3, 4, 5]

def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial (n - 1)

print ("Factorial of 5 is:", factorial(5))
print ("Factorial of 1 is:", factorial(0))

# Recursion function needs a base case because it knows where to stop
# Without a base case, it keeps calling itself and raises a Recursion error.

# Part 2: Lambda
double = lambda x: x * x
print (f"Double of 7 -> {double(7)}")

nums = [1, 2, 3, 4, 5]

double_nums = list(map(lambda x: x * 2, nums))
print ("Double nums ->", double_nums)

````

## Problem 4 — Pattern Matching with Regular Expressions  *(10 points)*

Use the `re` module to validate and extract text patterns.

> **Note:** The email pattern below is a deliberately *simplified* teaching pattern, not a real-world email validator. Grade against the expected output shown, not against a stricter pattern.

### What to do
1. Import the `re` module.
2. **Email check.** A list is provided:
   ```python
   emails = ["maria@unt.edu", "bad-email", "devon@my.unt.edu", "no_at_symbol.com", "aisha@gmail.com"]
   ```
   Loop through it and, using `re.match()` with the pattern `r"^[\w.]+@[\w.]+\.\w+$"`, print whether each address is `Valid` or `Invalid`, like:
   ```
   maria@unt.edu -> Valid
   bad-email -> Invalid
   ```
3. **Number extraction.** A string is provided:
   ```python
   text = "Order 3 lattes and 2 muffins for 21 dollars"
   ```
   Use `re.findall(r"\d+", text)` to pull out all the numbers, store them in `numbers`, and print the list (expected: `['3', '2', '21']`).
4. Add a **comment**: in the pattern `\d+`, what do `\d` and `+` each mean? (One short sentence.)

### Hints
- `re.match(pattern, text)` returns a match object (truthy) or `None` (falsy), so it works directly in an `if`.
- `\d` matches a digit; `+` means "one or more." Always write regex patterns as raw strings (the `r"..."` form).

````python
import re
emails = ["maria@unt.edu", "bad-email", "devon@my.unt.edu", "no_at_symbol.com", "aisha@gmail.com"]

email_format = r"^[\w.]+@[\w.]+\.\w+$"

#Printing invalid, valid emails
for email in emails:
  if re.match(email_format, email):
    print (email, "-> Valid")

  else:
    print(email, "-> Invalid")

print ("-" * 25)

text = "Order 3 lattes and 2 muffins for 21 dollars"
orders = re.findall(r"\d+", text)

print("Numbers found in text are:", orders)

#\d+ : means more than one digit
#\d : means a didgit
# + : means one or more
````

## Problem 5 — `datetime`, `repr()`, and Safe Use of `eval()` / `exec()`  *(10 points)*

Import a module from Python's standard library and use it for real work.

> **Point split (10 pts):** `datetime` import + two fixed dates (3) · day-of-week and day-difference output (3) · `repr()` vs `str()` with comment (2) · `eval()`/`exec()` safety comment (2).

### What to do
1. Import the `datetime` module.
2. Create a fixed date for the assignment due day:
   ```python
   due_date = datetime.date(2026, 7, 3)
   ```
   *(Use this fixed date — not `today()` — so your output is the same every time it's graded.)*
3. Create a second fixed date `start_date = datetime.date(2026, 6, 29)` (the release day).
4. Print, with labels:
   - the **day of the week** the assignment is due, using `due_date.strftime("%A")` (expected: `Friday`),
   - the number of **days between** start and due dates: `(due_date - start_date).days` (expected: `4`).
5. **`repr()` vs `str()`.** A string is provided in the starter code: `sample = "Line1\tLine2"`. Print both `str(sample)` and `repr(sample)` with labels, and add a **comment** describing the difference you see in the output.
6. Add one short **comment** explaining why `eval()` and `exec()` should not be used with untrusted user input. Do **not** use `eval()` or `exec()` in your code for this problem.

### Hints
- `datetime.date(year, month, day)` builds a date.
- Subtracting two dates gives a `timedelta`; its `.days` attribute is the whole-day count.
- `str(sample)` shows the text with the actual tab spacing. `repr(sample)` shows the escape character `\t`.
- `eval()` and `exec()` can run Python code, so they are unsafe with input you do not trust.


````python
# Name: Fathima Mohammed
# Problem 5: This program is about the datetime module + repr() vs str() + eval()/exec() risks.

import datetime

due_date = datetime.date(2026, 7, 3)
start_date = datetime.date(2026, 6, 29)

day_week = due_date.strftime("%A")
print("Day of the week is:", day_week)

days_between = (due_date - start_date).days
print ("Days between start and due dates is:", days_between)

print ("-" * 25)

sample = "Line1\tLine2"

print("str:", str(sample)) #output text in readable way
print("repr:", repr(sample)) #outputs text exactly given in the code.

# eval() and exec() should not be used because they can run code from a string,
# which can be unsafe.


````
