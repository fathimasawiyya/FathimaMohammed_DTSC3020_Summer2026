## Part 2 — The Receipt & Loyalty Stars  *(25 points)*

A customer orders **3 lattes** ($4.50 each) and **2 muffins** ($3.25 each). Texas sales tax is **8.25%**. The shop's loyalty program gives **1 star for every whole $5 spent** (based on the total).

### What to do
1. Store the four order values in variables: `latte_price`, `latte_qty`, `muffin_price`, `muffin_qty`.
2. Calculate, using **arithmetic operators**:
   - `subtotal` — price × quantity for both items, added together
   - `tax` — `subtotal * 8.25 / 100`
   - `total` — subtotal plus tax
3. Print the receipt using **f-strings with 2 decimal places**, like this:

   ```
   Subtotal: $20.00
   Tax (8.25%): $1.65
   Total: $21.65
   ```

   > 💡 **Money formatting recipe:** `f"${subtotal:.2f}"` prints a float with exactly 2 decimals.
4. Loyalty stars: first convert the total to whole dollars with `int(total)` (store it in `total_whole`). Then use **floor division and modulo**:
   - `stars_earned = total_whole // 5`
   - `toward_next = total_whole % 5`
   
   Print both with labels, e.g. `Stars earned: 4` and `Dollars toward next star: 1`.
5. The customer brought a **$25.00 budget**. Compute and print how much money they have left, formatted as money.
6. Add a **comment** answering: why do `//` and `%` together "split" the dollars into stars and leftover? (One sentence — hint: `21 = 4 * 5 + 1`.)


````python
# Fathima Mohammed
# Part 02: This part calculates the customer's receipt and loyalty stars.

# Storing the four order values in variables:
latte_price = 4.50
latte_qty = 3
muffin_price = 3.25
muffin_qty = 2

# Arithmetic Operators:

# Calculating subtotal:
subtotal = (latte_price * latte_qty) + (muffin_price * muffin_qty)
print (f"Subtotal: ${subtotal:.2f}")

# Calculating tax:
tax = subtotal * 8.25 / 100
print (f"Tax (8.25%): ${tax:.2f}")

# Calculating Total:
total = subtotal + tax
print (f"Total: ${total:.2f}")

# Conversion of decimal point to whole numbers:
total_whole = int(total)

# Floor division and modula calculation:
stars_earned = total_whole // 5 #floor division
toward_next = total_whole % 5  #modula

print (f"Stars earned: {stars_earned}")
print (f"Dollars toward next star: {toward_next}")

# //: finds how many complete $5 groups.
# %: find the leftover dollars after those groups are removed.

# Budget Calculations:
budget = 25.00
money_left = budget - total

print (f"Money left: ${money_left:.2f}")
````

````python
# Fathima Mohammed
# Part 05: This part is about running the weekly prize drawing.

import random
random.seed(23) #seed is set to give same output

# No the winner do not change because seed() is used.
# Seed gives the same output each time when the program runs.
# without seed, the output gives different winner and number each time.

customers = ["Maria", "Devon", "Aisha", "Carlos", "Lena", "Tyler"]
songs = ["Jazz Vibes", "Lo-Fi Beats", "Morning Acoustic", "Indie Mix", "Soft Piano"]

# Random numbers:
print ("The week's lucky number is:", random.randint(1,50)) # give random number between 1 and 50
print ("The discount roll is:", random.random()) #gives random number between 0 and 1

# Choice:
print ("The free drink winner is: ", random.choice(customers))

#Sample
print ("The two different coupon winners are: ", random.sample(customers, 2))

# Shuffle:
random.shuffle(songs)
print ("The new playlist order is:", songs)




````
